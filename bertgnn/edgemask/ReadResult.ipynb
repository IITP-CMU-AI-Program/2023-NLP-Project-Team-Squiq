```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
loss_list = []
acc_list = []

with open("log.txt", "r") as f:
    for line in f.readlines():
        loss_str, acc_str = line.rstrip().split("\t")
        loss_list.append(float(loss_str))
        acc_list.append(float(acc_str))
```

```python
plt.rcParams.update({"font.size": 22})

plt.figure(figsize=(5, 5), dpi=100)
plt.plot(range(len(loss_list)), loss_list, linewidth=10)
plt.xlabel("Training Step")
plt.ylabel("Train Loss")
plt.savefig("train_loss.jpg", bbox_inches="tight")
```

```python
plt.rcParams.update({"font.size": 22})

plt.figure(figsize=(5, 5), dpi=100)
plt.plot(range(len(acc_list)), acc_list, linewidth=10)
plt.xlabel("Training Step")
plt.ylabel("Train ACC")
plt.savefig("train_acc.jpg", bbox_inches="tight")
```
