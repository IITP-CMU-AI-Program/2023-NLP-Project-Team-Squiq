```python
import os
```

```python
train_loss_list = []
test_loss_list = []

with open("log.txt", "r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        train_loss = float(line.rstrip().split(" ")[-1])
        line = f.readline()
        test_loss = float(line.rstrip().split(" ")[-1])

        train_loss_list.append(train_loss)
        test_loss_list.append(test_loss)
```

```python
import matplotlib.pyplot as plt

plt.rcParams.update({"font.size": 22})

plt.figure(figsize=(5, 5), dpi=100)
plt.plot(range(len(train_loss_list)), train_loss_list, linewidth=10)
plt.xlabel("Training Step")
plt.ylabel("Train Loss")
plt.savefig("train_loss.jpg", bbox_inches="tight")
```

```python
import matplotlib.pyplot as plt

plt.rcParams.update({"font.size": 22})

plt.figure(figsize=(5, 5), dpi=100)
plt.plot(range(len(test_loss_list)), test_loss_list, linewidth=10)
plt.xlabel("Training Step")
plt.ylabel("Test Loss")
plt.savefig("test_loss.jpg", bbox_inches="tight")
```
